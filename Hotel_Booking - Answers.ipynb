# Data Cleaning

```python
# Rename columns to lower letters
hotel = hotel.rename(columns=str.lower)
hotel
```

```python
# show hotel columns
hotel.columns
```

```python
# Drop some features
hotel = hotel[['hotel','arrival_date_year','stays_in_week_nights','adults','children','babies','is_repeated_guest','previous_cancellations','previous_bookings_not_canceled','is_canceled' ]]
hotel
```

```python
# Drop some features
hotel = hotel.drop(['arrival_date_day_of_month', 'deposit_type', 'country', 'customer_type', 'stays_in_weekend_nights', 'meal', 'stays_in_weekend_nights', 'market_segment','distribution_channel','reserved_room_type','assigned_room_type','lead_time','arrival_date_month', 'adr', 'required_car_parking_spaces','total_of_special_requests','arrival_date_week_number','days_in_waiting_list','company','agent', 'booking_changes',], axis=1)
hotel.head(10)
```

```python
# Missing values in perecentege
hotel.isna().mean()
```

```python
# Missing values in Values
hotel.isnull().sum()
```

```python
# Show the rows with missing values
hotel.loc[hotel.children.isnull()]
```

```python
# In column "children", we'll put 0 for 'Nan'
hotel.loc[hotel.children.isnull(), 'children'] = 0
hotel.isnull().sum()
```

No more missing values here !!!

```python
# Change categorical values to "one hot"
hotel = pd.get_dummies(hotel)
hotel.head(10)
```

```python
# Check all values are indeed numeric
hotel.dtypes
```

All features are numeric, perfect.

```python
# Plot correlations to is_canceled
hotel.corr()['is_canceled'].drop('is_canceled').plot(kind='barh')
```

let's check "previous_cancelltaions"

```python
# What is the "previous_cancellations" distribution?
previous_cancellations_distribution = hotel.previous_cancellations.value_counts(normalize=True).sort_index()
previous_cancellations_distribution
```

```python
# Too many values, let's group them together
# Add a column "pre_cancellation_qty" with the folowing values:
    # '0'
    # '1-5'
    # '6-10'
    # '11+'


hotel['pre_cancellation_qty'] = None


hotel.loc[(hotel['previous_cancellations'] == 0),'pre_cancellation_qty'] = '0'
hotel.loc[(hotel['previous_cancellations'] > 0) & (hotel['previous_cancellations'] <= 5),'pre_cancellation_qty'] = '1-5'
hotel.loc[(hotel['previous_cancellations'] > 5) & (hotel['previous_cancellations'] <= 10),'pre_cancellation_qty'] = '6-10'
hotel.loc[(hotel['previous_cancellations'] > 10),'pre_cancellation_qty'] = '11+'

hotel.head(1000)
```

```python
# What is the "pre_cancellation_qty" distribution?
pre_cancellation_qty_distribution = hotel.pre_cancellation_qty.value_counts(normalize=True).sort_index()
pre_cancellation_qty_distribution
```

```python
# What are the chances to cancel for each feature value?
pre_cancellation_qty_distribution_Chances = hotel.groupby('pre_cancellation_qty').is_canceled.mean()
pre_cancellation_qty_distribution_Chances
```

```python
# Let's plot it all into one graph

pre_cancel_stats = pd.DataFrame({'pre_cancellation_dis':pre_cancellation_qty_distribution, 'cancelation_dist':pre_cancellation_qty_distribution_Chances})

pre_cancel_stats.plot(kind='bar')
plt.xlabel('pre_cancellation_qty_distribution')
plt.ylabel('pre_cancellation_qty_distribution_Chances')
plt.ylim(0, 1)
plt.title('');
```

Intresting, if you canceled between 1-5 you have more chances to cancel than people who canceld 6-10
